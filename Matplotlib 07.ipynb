```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

### Sample Data ##

```python
data_uas=[
    ['Bejo', 70],
    ['Tejo', 83],
    ['Cecep', 62],
    ['Wati', 74],
    ['Karti', 71]
] # membuat list 2 rangkap
```

```python
data_uas # untuk menampilkan data yang tertampung dalam data uas
```

### Table Plot Dengan 00 Style ##

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')
table.set_fontsize(14) # pengaturan font size
table.scale(1, 4) # ukuran kolom, baris

ax.axis(False)

plt.show()
```

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')
table.set_fontsize(14) # pengaturan font size
table.scale(0.5, 6) # ukuran kolom lebih kecil, baris lebih lebar

ax.axis(False) #menghilangkan sumbu

plt.show()
```

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas,
                 loc='center')
table.set_fontsize(14) # pengaturan font size
table.scale(0.5, 3) # ukuran kolom sama, baris lebih pendek

#ax.axis(False) 

plt.show()
```

## Table Plot Dengan Pyplot Style ###

```python
table = plt.table(cellText=data_uas,
                 loc='center')

table.set_fontsize(14) # pengaturan font size
table.scale(1, 4) # ukuran kolom, baris

ax = plt.gca() #get current access
ax.axis(False) #bagian dari akses

plt.show()
```
