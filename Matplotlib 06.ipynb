### Import Modules ##

```python
%matplotlib inline
```

```python
import matplotlib
import matplotlib.pyplot as plt

print(matplotlib.__version__)
```

### Sample Pie Plot  ##

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
plt.pie(data,
#      labels=kategori, optional
       autopct='%1.1f%%',
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
plt.pie(data,
#      labels=kategori, optional jadi tidak ada keterangan per bagian
#      autopct='%1.1f%%', optional jadi tidak ada presentase saja bedanya
       startangle=90)

plt.title('Contoh Pie Plot')

plt.show()
```

```python
plt.pie(data,
       labels=kategori,
       autopct='%1.1f%%',
       startangle=180) # berubah putaran

plt.title('Contoh Pie Plot')

plt.show()
```

### Pengaturan Legend Dan Color ###

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data,
        labels=kategori, # untuk memunculkan keterangan per bagian di data
        colors=warna, # untuk memunculkan warna sesuai data
        autopct='%1.1f%%', #untuk memunculkan presentase
        startangle=90) # arahnya skrng berlawanan dengan jarum jam 
# startangle untuk menampilkan data sesuai sudut yang di tentukan

plt.legend() # keterangan plotting
plt.title('Contoh Pie Plot')

plt.show()
```

```python
plt.pie(data,
        labels=kategori, # untuk memunculkan keterangan per bagian di data
        colors=warna, # untuk memunculkan warna sesuai data
        autopct='%1.1f%%', #untuk memunculkan presentase
        startangle=60) # jadi searah jarum jam

#plt.legend() # keterangan plotting
plt.title('Contoh Pie Plot')

plt.show()
```

### Exploded Pie Plot ##

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08] #semakin besar semakin jauh jarak ledakannya
```

```python
plt.pie(data,
        labels=kategori, # untuk memunculkan keterangan per bagian di data
        colors=warna, # untuk memunculkan warna sesuai data
        explode=explode_var, #untuk memnetukan bagian yang explode/meledak 
        autopct='%1.1f%%', #untuk memunculkan presentase
        startangle=90) # untuk menampilkan data sesuai sudut yang di tentukan

plt.title('Contoh Exploded Pie Plot')

plt.show()
```
