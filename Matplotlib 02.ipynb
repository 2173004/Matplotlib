## Pengaturan Format Pada Plot ##

```python
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], 'g*')
plt.axis([0, 6, 0, 20])
plt.show()
```

### Multi Plot Dalam Satu Pemanggilan Fungsi plot() ##

```python
t = np.arange(0., 5., 0.2)
t
```

plt.plot(t, t, 'r--')
plt.plot(t, t**2, 'bs')

plt.show()

```python
plt.plot(t, t, 'r--',
        t, t**2, 'bs',
        t, t**3, 'g^')
plt.show()
```

## Plotting Dengan Keywords

```python
data = {'a': np.arange(50),
       'c': np.random.randint(1, 50, 50),
       'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

data
```

Menampilkan data dictionary

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()b
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```

### Plotting Untuk Tipe Data Categorical

Matplotlib juga mendukung plotting untuk data dengan tipe katergori atau biasa dikenal dengan istilah cageorical data atau categorical variable

```python
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]
```

```python
plt.plot(names, values)
plt.show()
```

```python
plt.scatter(names, values)
```

```python
plt.bar(names, values)
plt.show()
```
