```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

print(matplotlib.__version__)
print(np.__version__)
```

## Simple Dataset ##

```python
data = {'Item A': 109438,
        'Item B': 103569,
        'Item C': 112214,
        'Item D': 112591,
        'Item E': 100934,
        'Item F': 103660,
        'Item G': 137351,
        'Item H': 123381,
        'Item I': 135841,
        'Item J': 104437}

data
```

```python
items = tuple(data.keys()) #mengakses keynya saja
items
```

```python
count = tuple(data.values()) #mengakses nilainya saja
count
```

## Simple Plot ##

```python
fig, ax = plt.subplots()

ax.barh(items, count) #plotting horizontal

plt.show()
```

## Pengaturan Style ##

```python
plt.style.available #melihat style yang tersedia
```

```python
plt.style.use('bmh') #menggunakan salah satu style yang ada

fig, ax = plt.subplots()
ax.barh(items, count)
plt.show()
```

### Pengaturan Tick Label ###

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels() #untuk mendapatkan akses yang ada

#set properties
plt.setp(labels,
         rotation=45, #dirotate 45 derajat berlawanan dengan arah jarum jam
         horizontalalignment='right') #perataan kanan

plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

labels = ax.get_xticklabels() #untuk mendapatkan akses yang ada

#set properties
plt.setp(labels,
         rotation=90, #dirotate 90 derajat berlawanan dengan arah jarum jam
         horizontalalignment='left') #perataan kiri

plt.show()
```

### Pengaturan Format Pada Ticker ##

```python
from matplotlib.ticker import FuncFormatter
```

```python
def ribuan(x, pos):
    return f'{int(x/1000)}k'
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

formatter = FuncFormatter(ribuan)
ax.xaxis.set_major_formatter(formatter)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

### Pengaturan Label Pada Sumbu (Axis) dan Judul (Title) ###

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

## Penambahan Garis (Vertical/Horisontal Line) Pada Plot ##

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count), #garis vertical pada akses
           ls='--', #line style
           color='r') #warna garis

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(80000,
           ls='--',
           color='r')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

plt.show()
```

### Menyimpan Hasil Plot Ke Dalam Suatu File ###

```python
fig.canvas.get_supported_filetypes()
```

```python
fig, ax = plt.subplots()
ax.barh(items, count)

ax.axvline(np.mean(count),
           ls='--',
           color='r')

ax.set(title='Contoh Plot',
       xlabel='Jumlah',
       ylabel='Items')

#untuk save hasil ploting
fig.savefig('sales.png', #nama file
            transparent=False, #transparasi
            dpi=80) #resolusi
```
